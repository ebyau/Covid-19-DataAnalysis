```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
corona_dataset_csv = pd.read_csv("data/covid19_confirmed_global.csv")
```

```python
corona_dataset_csv.head()
```

```python
corona_dataset_csv.shape
```

```python
df = corona_dataset_csv.drop(["Lat","Long"],axis=1) # axis =1 selescts to drop the cols
df.head()
```

```python
corona_dataset_csv.drop(["Long","Lat"],axis=1,inplace=True) # inplace modifies the current dataset no need to save it in another file
corona_dataset_csv
```

```python
# sums up all cases for a given country 
corona_dataset_aggregated = corona_dataset_csv.groupby("Country/Region").sum() 
```

```python
corona_dataset_aggregated.head()
```

```python
print("Aggregated shape",corona_dataset_aggregated.shape)
print("Original shape", corona_dataset_csv.shape)
```

```python
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["Italy"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
plt.legend()
```

```python
# infections over 3 days
corona_dataset_aggregated.loc["China"][:3].plot() 
```

```python
# plot shows the infection rate on daily basis
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
# the max infection rate
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
print("Italy maximum infection rate ", corona_dataset_aggregated.loc["Italy"].diff().max())
print("Spain maximum infection rate ",corona_dataset_aggregated.loc["Spain"].diff().max())
```

```python
#maximum infection rate for all countries
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["Max_infection_rates"] = max_infection_rates    
```

```python
corona_dataset_aggregated.tail()
```

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["Max_infection_rates"])
#corona_data = corona_dataset_aggregated["Max_infection_rates"]
corona_data.head()
```

```python
hapiness_report = pd.read_csv("data/World_happiness_report_2020.csv")
```

```python
hapiness_report.head()
```

```python
#delete useless cols
useless_cols = ["Regional indicator","Ladder score","Standard error of ladder score",
               "upperwhisker","lowerwhisker","Generosity","Perceptions of corruption",
               "Ladder score in Dystopia","Explained by: Log GDP per capita",
               
                "Dystopia + residual","Explained by: Perceptions of corruption",
               "Explained by: Generosity","Explained by: Freedom to make life choices",
               "Explained by: Healthy life expectancy","Explained by: Social support"]
        
```

```python
hapiness_report.drop(useless_cols,axis=1,inplace=True)
hapiness_report.head()
```

```python

hapiness_report.set_index("Country name",inplace=True)
```
